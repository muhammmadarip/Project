```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("global firepower 2022 wide.csv")
df.head(11)
```

```python
df.info()
```

```python
df.describe().round()
```

```python
del df["Waterways (usable)"]
del df["Square Land Area"]
del df["Dedicated Attack"]
del df["Coastline Coverage"]
del df["Trainers"]
df
```

```python
df.corr
```

```python
defense_budget = df.nlargest(15,'Defense Budget')
defense_budget
```

```python
plt.figure(figsize = (10,5))
sns.barplot(y='country', x='Defense Budget',data=df.sort_values('Defense Budget', ascending=False).head(15))
plt.show()
```

```python
defense_budget = df.nlargest(15,'Active Personnel')
defense_budget
```

```python
plt.figure(figsize = (10,5))
sns.barplot(y='country', x='Active Personne',data=df.sort_values('Active Personnel', ascending=False).head(15))
plt.show()
```

```python
plt.figure(figsize = (6,6))
sns.scatterplot(x='Active Personnel', y='Defense Budget', hue='country',data=df.sort_values('Defense Budget', ascending=False).head(15))
plt.legend(loc = 2, bbox_to_anchor = (1,1))
plt.show()
```

- Berdasarkan analisis diatas, negara dengan anggaran militer tertinggi adalah US sedangkan negara dengan jumlah personil militer terbanyak adalah china. Anggaran militer suatu negara tidak dipengaruhi oleh jumlah personil aktif.
